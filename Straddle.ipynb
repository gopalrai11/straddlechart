```python
import time
import matplotlib.pyplot as plt
```

```python
t = time.localtime()
current_time = time.strftime("%H:%M:%S", t)
print(current_time)
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,title='straddle chart')
fig.show()
```

```python

i = 9.15
x, y = [], []

while True:
    hdfc=nse.get_quote('NIFTY', segment=Segment.OPT, optionType=OptionType.CE, strike=18000)
    hdfc2=pd.DataFrame(hdfc,[18000])

    hdfc=nse.get_quote('NIFTY', segment=Segment.OPT, optionType=OptionType.PE, strike=18000)
    hdfc1=pd.DataFrame(hdfc,[18000])


    total=(hdfc1['lastPrice'] + hdfc2['lastPrice'])
    total.head()
    x.append(i)
    y.append(total.head())
    
    ax.plot(x, y, color='b')
    
    fig.canvas.draw()
    
    ax.set_xlim(left=max(0, i-50), right=i+50)
    
    time.sleep(180)
    i += 1
```
